# Basic Descriptive Statistics

Create a list of 10 test scores 100,90,90,84,87,75,73,64,65,72

Perform the following analysis on the collection using Python built in functions or code you write yourself.

```python
scores = [100, 90, 90, 84, 87, 75, 73, 64, 65, 72]
scores.sort()
print(f"Scores: {scores}")
```

## 1. Measures of Central Tendancy

Find the mean, median, and mode of the collection.

Which statistic is more distorted by outliers?

```python
mean = sum(scores) / len(scores)
print(f"Mean: {mean}")

median = 0
if len(scores) % 2 == 0:
  median = (scores[int((len(scores) / 2) + 0.5)] + scores[int(len(scores) / 2 - 0.5)]) / 2
else:
  median = scores[len(scores) / 2]
print(f"Median: {median}")

counts = {}
mode = 0;
for score in scores:
  if not (score in counts):
    counts[score] = 1
  else:
    counts[score] += 1
highest = 0
for key in counts.keys():
  if counts[key] > highest:
    highest = counts[key]
for key in counts:
  if counts[key] == highest:
    mode = key
print(f"Mode: {mode}")
```

## 2. Measures of Variability

Find the _range, standard deviation, and variance of the collection.

What is a drawback of the _range?

```python
range_of_set = scores[-1] - scores[0]
print(f"Range: {range_of_set}")

deviations = scores
print(deviations)
for i in range(len(deviations)):
  deviations[i] = (deviations[i] - mean) ** 2
standard_deviation = (sum(deviations) / len(deviations)) ** 0.5
print(f"Standard deviation: {standard_deviation}")

squared_differences = []
for score in scores:
  squared_differences.append((score - mean) ** 2)
variance = sum(squared_differences) / len(scores)
print(f"Variance: {variance}")
```

## 3. Measure of Frequency

Count the items in the collection.

```python
print(f"Number of elements: {len(scores)}")
print(f"Number of unique elements: {len(set(scores))}")
```
