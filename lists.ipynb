# Lists

### Example #1

```python
my_list = ["William",20,189.5]                      # Able to store heterogeneous data
my_list
```

### Example #2

```python
my_list[0]                                          # Subsetable
```

### Example #3

```python
my_list[0] = "Roberto"                              # Mutable (reassign a value)
my_list
```

### Example #4

```python
my_list.append("Li")                                # Mutable (Lists are objects, so they have methods)
my_list
```

### Example #5

```python
my_list.append("19")
my_list
```

### Example #6

```python
my_list.append(150.5)
my_list
```
